```python
import pandas as pd
import statsmodels.formula.api as smf

# --- Load dataset with cluster labels ---
df = pd.read_csv("q2_outputs/Spotify_Q2_with_clusters.csv")

# --- Inspect key variables ---
print(df[["Cluster", "Cluster_Name", "log_Views", "log_Likes", "log_Comments", "Album_type"]].head())

# --- Baseline regression: cluster effect on Views ---
model1 = smf.ols("log_Views ~ C(Cluster)", data=df).fit()
print("\n=== Model 1: Cluster only (Views) ===")
print(model1.summary())

# --- Add Album_type as control variable ---
model2 = smf.ols("log_Views ~ C(Cluster) + C(Album_type)", data=df).fit()
print("\n=== Model 2: Cluster + Album_type (Views) ===")
print(model2.summary())

# --- Full model with interaction term ---
model3 = smf.ols("log_Views ~ C(Cluster) * C(Album_type)", data=df).fit()
print("\n=== Model 3: Interaction (Views) ===")
print(model3.summary())

# --- Repeat analysis for Likes ---
model_likes = smf.ols("log_Likes ~ C(Cluster) * C(Album_type)", data=df).fit()
print("\n=== Model: Interaction (Likes) ===")
print(model_likes.summary())

# --- Repeat analysis for Comments ---
model_comments = smf.ols("log_Comments ~ C(Cluster) * C(Album_type)", data=df).fit()
print("\n=== Model: Interaction (Comments) ===")
print(model_comments.summary())
```
